# 不平衡数据处理措施
**数据指标**：ACC、F1、Recall、Confusion Matrix

**测试实验**：

1、**实验数据集以及模型**：CIFAR-10数据集，模型采用Renet50、ViT-H/14

2、**实验测试方法**：损失函数策略选择测试loss分为focal loss、交叉熵损失、先验损失等、数据策略测试重采用欠采样多种

3、**构建不平衡标签分布**，数据集标签分布为Airplanes、Automobiles、Birds、Cats、Deer、Dogs、Frogs、Horses、Ships、Trucks，**不平衡方式构建如下**：$N_i=N_{\max}\times \mu^{\frac{i}{C-1}}$，其中 $\mu$表示不平衡因子、$C$表示类别、$N$表示整体数据数量

4、**构建错误标签**：随机选择部分相似标签进行转换，Cat-Dog）、Automobile-Truck、Deer-Horse、Bird-Airplane
