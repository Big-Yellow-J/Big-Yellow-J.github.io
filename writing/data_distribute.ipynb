# 常用分类指标及其计算方式
> 分类任务中除去计算固定的几个指标外，**一定一定要去输出最后模型预测的混淆矩阵**（只去看指标不够直观，往往需要去看模型“侧重点”）

在分类问题中，通常会先构建一个**混淆矩阵（Confusion Matrix）**，绝大多数分类指标都基于它计算。以二分类问题为例：

|    实际\预测     |    正类（Positive）    |    负类（Negative）    |
| :----------: | :----------------: | :----------------: |
| 正类（Positive） | TP（True Positive）  | FN（False Negative） |
| 负类（Negative） | FP（False Positive） | TN（True Negative）  |
- TP（真正例）：实际是正样本，预测也是正样本
- TN（真负例）：实际是负样本，预测也是负样本
- FP（假正例）：实际是负样本，预测成正样本（误报）
- FN（假负例）：实际是正样本，预测成负样本（漏报）
> 直接去记字母，比如说T就是预测正常，P和N分别表示正类和负类
## 数值指标
几类常见的数值指标计算方式总结如下：

|       指标       |                            计算方式                            |          通俗解释           |
| :------------: | :--------------------------------------------------------: |:---------------------------:|
|    准确率（ACC）    |                  $(TP+TN)/(TP+TN+FP+FN)$                   |  答对内容占所有回答中比例   |
| 精确率（Precision） |                        $TP/(TP+FP)$                        | 预测正类中模型有多少是真的  |
|  召回率（ReCall）   |                        $TP/(TP+FN)$                        | 实际正类中模型可以预测多少  |
|       F1       | $\frac{2\times Precision\times Recall}{Precision+ Recall}$ |                             |
|      UAR       |            $\frac{1}{K} \sum^{K}_{i=i}Recall_i$            |                             |
> 精确率和召回率关注点，**前者关注从预测的正类中可以正确判别多少，后者关注实际正类中可以预测出多少真正为正类比例**

# 不平衡数据不同处理措施实验

## 实验设置
> 所有测试代码 data_distribute.py

**数据指标**：ACC、F1、Recall、Confusion Matrix

**测试实验**：

1、**实验数据集以及模型**：CIFAR-10数据集，模型采用Renet50、ViT-H/14

2、**实验测试方法**：损失函数策略选择测试loss分为focal loss、交叉熵损失、先验损失等、数据策略测试重采用欠采样多种

3、**构建不平衡标签分布**，数据集标签分布为Airplanes、Automobiles、Birds、Cats、Deer、Dogs、Frogs、Horses、Ships、Trucks，**不平衡方式构建如下**：$N_i=N_{\max}\times \mu^{\frac{i}{C-1}}$，其中 $\mu$表示不平衡因子、$C$表示类别、$N$表示整体数据数量

4、**构建错误标签**：随机选择部分相似标签进行转换，Cat-Dog、Automobile-Truck、Deer-Horse、Bird-Airplane

## 改进措施
1、对于数据集进行多种采用策略（欠采样、重采用等）；2、对于loss进行改进（交叉熵损失、focal loss等）；3、改进模型输出策略，比如说对于大类标签直接控制模型要求输出>0.5

## 结论
> 与其说是结论不如说是 “条件结论成立” 以为每个实验所进行的数据都是不同的，只能在 “限定范围”内有效

部分结论来自[^1]中的描述结果：**1、单纯重采样/重加权方法的实际收益有限，甚至可能在高IR场景中适得其反**。这是以为重采样容易引入低质量样本、在不平衡比（最多样本数量/最少样本数量）>10的时候欠采样容易丢失关键信息；**2、集成策略在性能和鲁棒性上都有显著优势，适合大多数不平衡场景**，不过考虑深度学习中反而用到比较少（因为引入新的模型疑问引入新的计算，导致计算上效率不高），可以直接考虑使用 “专家策略”，同一个模型去选择冻结其特征提取部分，重新去训练分类器而后去做不同类微调，比如分类头A做3类预测，分类头B做另外3类预测。
